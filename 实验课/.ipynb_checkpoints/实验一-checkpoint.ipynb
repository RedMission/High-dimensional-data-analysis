根据泰勒展开式原理绘制：指数函数

```python
import numpy as np
import math

def calc_e_small(x):
    n = 10
    f = np.arange(1, n+1).cumprod()#阶乘
    b = np.array([x]*n).cumprod()# 算的是x的n次方
    return np.sum(b / f) + 1

def calc_e(x):
    reverse = False
    if x < 0:   # 处理负数
        x = -x
        reverse = True
    ln2 = 0.69314718055994530941723212145818
    c = x / ln2
    a = int(c+0.5)
    b = x - a*ln2
    y = (2 ** a) * calc_e_small(b)
    if reverse:
        return 1/y
    return y
```

```python
#根据泰勒展开式原理绘制：指数函数
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

if __name__ == "__main__":
    t1 = np.linspace(-2, 0, 10, endpoint=False) #linspace指定的间隔内返回均匀间隔的数字
    t2 = np.linspace(0, 3, 20)
    t = np.concatenate((t1, t2)) #concatenate：数组拼接
    print("t:",t)     # 横轴数据
    y = np.empty_like(t) #依据给定数组t的形状和类型返回一个新的空数组
    for i, x in enumerate(t): #enumerate() 函数用于将t返回 enumerate(枚举) 对象
        y[i] = calc_e(x)
        print("\nx:",x, "\ny[i]:",y[i],"\nmath.exp(x):", math.exp(x))
    
    #可视化   
    plt.figure(facecolor='w')
    mpl.rcParams['font.sans-serif'] = [u'SimHei'] #显示中文
    mpl.rcParams['axes.unicode_minus'] = False
    plt.plot(t, y, 'b-', t, y, 'go', linewidth=2)
    plt.title(u'Taylor展式的应用 - 指数函数', fontsize=18)
    plt.xlabel('X', fontsize=15)
    plt.ylabel('exp(X)', fontsize=15)
    plt.grid(True, ls=':')
    plt.show()
```

```python
#根据泰勒展开式原理绘制正弦函数
import numpy as np
import math
import matplotlib as mpl
import matplotlib.pyplot as plt


def calc_sin_small(x):
    x2 = -x ** 2
    t = x
    f = 1
    sum = 0
    for i in range(10):
        sum += t / f
        t *= x2
        f *= ((2*i+2)*(2*i+3))
    return sum


def calc_sin(x):
    a = x / (2*np.pi)
    k = np.floor(a)
    a = x - k*2*np.pi
    return calc_sin_small(a)


if __name__ == "__main__":
    t = np.linspace(-2*np.pi, 2*np.pi, 100, endpoint=False)
    print(t)     # 横轴数据
    y = np.empty_like(t)
    for i, x in enumerate(t):
        y[i] = calc_sin(x)
    mpl.rcParams['font.sans-serif'] = [u'SimHei']
    mpl.rcParams['axes.unicode_minus'] = False
    plt.figure(facecolor='w')
    plt.plot(t, y, 'r-', t, y, 'go', linewidth=2)
    plt.title(u'Taylor展式的应用 - 正弦函数', fontsize=18)
    plt.xlabel('X', fontsize=15)
    plt.ylabel('sin(X)', fontsize=15)
    plt.xlim((-7, 7))
    plt.ylim((-1.1, 1.1))
    plt.grid(True)
    plt.show()
```

2)	设计数据进行分析并可视化结果

```python
def TestData(size):
    np.random.seed(1)
    X = np.random.randint(1, 10, size)

    X=X.reshape(-1,3) #变换数组形状
    
    Y_or = X[:,2]  #保存原始第三列的值

    for i in range(X.shape[0]):#设定第三列小于3记为0，大于3小于6记为1，大于6记为2
        if X[i,2] <= 3:
            X[i,2] = 0
        elif X[i,2] <= 6:
            X[i,2] = 1
        else: X[i,2] = 2
    
    X_train = X[:,:2] #切分前2列为X
    Y_train = X[:,2]  #切分第3列为Y
    return(X_train,Y_or,Y_train)
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

def TestPlt(X_train):
    T = np.arctan2(X_train[:,0],X_train[:,1])
    plt.scatter(X_train[:,0],X_train[:,1],s=40,c=T,marker='.')     #散点图生成
    plt.xlabel('X1')                             # plt.xlabel 设置x轴标签
    plt.ylabel('X2')                             # plt.ylabel 设置y轴标签
    plt.title('Test')                                    #plt.title  设置图像标题

    plt.show()
    
def draw_3D(X_train,Y_or):
    ax = Axes3D(plt.figure())
    ax.scatter(X_train[:,0], X_train[:,1], Y_or)
    ax.set_title("Test")
    plt.show()
```

```python
if __name__ == "__main__":
    X_train,Y_or,Y_train = TestData(120)
    draw_3D(X_train,Y_or) #观察数据分布
    TestPlt(X_train)
```
