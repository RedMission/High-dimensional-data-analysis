## 1.PCA

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def pca(data, n_dim):
    '''
    :param data: (n_samples, n_features(D))é™ç»´çš„çŸ©é˜µ
    :param n_dim: target dimensionsç›®æ ‡ç»´åº¦
    :return: (n_samples, n_dim)
    '''
    data = data - np.mean(data, axis = 0, keepdims = True) #è¿›è¡Œé›¶å‡å€¼åŒ–
    cov = np.dot(data.T, data) #è®¡ç®—å½’ä¸€åŒ–åŽçš„æ•°æ®é›†çš„åæ–¹å·®çŸ©é˜µ
    eig_values, eig_vector = np.linalg.eig(cov) #åæ–¹å·®çŸ©é˜µçš„ç‰¹å¾å€¼åŠå¯¹åº”çš„ç‰¹å¾å‘é‡
    indexs_ = np.argsort(-eig_values)[:n_dim] #å–å‰n_dimè¡Œç»„æˆçŸ©é˜µï¼Œä¿ç•™æœ€å¤§çš„n_dimå„ç‰¹å¾å‘é‡
    picked_eig_values = eig_values[indexs_]
    picked_eig_vector = eig_vector[:, indexs_] #æ‰¾å‡ºn_dimä¸ªç‰¹å¾å€¼ç›¸åº”çš„ç‰¹å¾å‘é‡
    data_ndim = np.dot(data, picked_eig_vector) #ð‘Œ=picked_eig_vectorÂ·ð‘‹
    return data_ndim


# data é™ç»´çš„çŸ©é˜µ(n_samples, n_features)
# n_dim ç›®æ ‡ç»´åº¦
# fit n_features >> n_samples, reduce cal
def highdim_pca(data, n_dim):
    '''
    when n_features(D) >> n_samples(N), highdim_pca is O(N^3)
    :param data: (n_samples, n_features)
    :param n_dim: target dimensions
    :return: (n_samples, n_dim)
    '''
    N = data.shape[0]
    data = data - np.mean(data, axis = 0, keepdims = True)

    Ncov = np.dot(data, data.T)

    Neig_values, Neig_vector = np.linalg.eig(Ncov)
    indexs_ = np.argsort(-Neig_values)[:n_dim]
    Npicked_eig_values = Neig_values[indexs_]
    # print(Npicked_eig_values)
    Npicked_eig_vector = Neig_vector[:, indexs_]
    # print(Npicked_eig_vector.shape)

    picked_eig_vector = np.dot(data.T, Npicked_eig_vector)
    picked_eig_vector = picked_eig_vector/(N*Npicked_eig_values.reshape(-1, n_dim))**0.5
    # print(picked_eig_vector.shape)

    data_ndim = np.dot(data, picked_eig_vector)
    return data_ndim

```

```python
if __name__ == "__main__":
    data = load_iris()
    X = data.data
    Y = data.target
    data_2d1 = pca(X, 2) #åˆ©ç”¨pcaé™ä½Žä¸º2ç»´
    #å¯è§†åŒ–
    plt.figure(figsize=(14,5))
    plt.subplot(121)
    plt.title("demo_PCA")
    plt.scatter(data_2d1[:, 0], data_2d1[:, 1], c = Y)
    #ä¸Žsklearnä¸­PCAé™ç»´åšå¯¹æ¯”
    sklearn_pca = PCA(n_components=2)
    data_2d2 = sklearn_pca.fit_transform(X)
    plt.subplot(122)
    plt.title("sklearn_PCA")
    plt.scatter(data_2d2[:, 0], data_2d2[:, 1], c = Y)
    plt.show()
    
```

## 2.PCR

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def simple_regression(X,y):
    # Create linear regression object
    regr = linear_model.LinearRegression()
    # Fit
    regr.fit(X, y)
    # Calibrationæ ¡å‡†
    y_c = regr.predict(X)
    # Cross-validationäº¤å‰éªŒè¯
    y_cv = cross_val_predict(regr, X, y, cv=10)
    # Calculate scores for calibration and cross-validation
    #è®¡ç®—åˆ†æ•°ä»¥è¿›è¡Œæ ¡å‡†å’Œäº¤å‰éªŒè¯
    score_c = r2_score(y, y_c)
    score_cv = r2_score(y, y_cv)
    # Calculate mean square error for calibration and cross validation
    #è®¡ç®—å‡æ–¹è¯¯å·®ä»¥è¿›è¡Œæ ¡å‡†å’Œäº¤å‰éªŒè¯
    mse_c = mean_squared_error(y, y_c)
    mse_cv = mean_squared_error(y, y_cv)
    return(y_cv, score_c, score_cv, mse_c, mse_cv)
```

```python
def pcr(X,y,pc):#pc:è¦é€‰æ‹©çš„æœ€å¤§ä¸»æˆåˆ†æ•°
    ''' Step 1: PCA on input data'''
    # Define the PCA object
    pca = PCA()
    # Preprocess (2) Standardize features by removing the mean and scaling to unit variance
    #é¢„å¤„ç†ï¼ˆ2ï¼‰é€šè¿‡åŽ»é™¤å‡å€¼å¹¶ç¼©æ”¾åˆ°å•ä½æ–¹å·®æ¥æ ‡å‡†åŒ–ç‰¹å¾
    Xstd = StandardScaler().fit_transform(X)
    # Run PCA producing the reduced variable Xred and select the first pc components
    #è¿è¡ŒPCAäº§ç”Ÿå‡å°‘çš„å˜é‡Xredå¹¶é€‰æ‹©ç¬¬ä¸€ä¸ªpcç»„ä»¶
    Xreg = pca.fit_transform(Xstd)[:,:pc]
    ''' Step 2: regression on selected principal components'''
    '''æ­¥éª¤2ï¼šå¯¹é€‰å®šçš„ä¸»è¦æˆåˆ†è¿›è¡Œå›žå½’'''
    y_cv, score_c, score_cv, mse_c, mse_cv = simple_regression(Xreg, y)
    return(y_cv, score_c, score_cv, mse_c, mse_cv)
```

```python
# Read data
data = pd.read_csv(r'E:\Documents\DAY\DA_python\data\nirpyresearch-master\data\peach_spectra_brixvalues.csv')
X = data.values[:,1:]
y = data['Brix']

npc = 40 # maximum number of principal components
pc = range(1,npc+1,1)

# Define arrays for R^2 and MSE
r2c = np.zeros(npc)
r2cv = np.zeros(npc)
msec = np.zeros(npc)
msecv = np.zeros(npc)
for i in pc:
    predicted, r2c[i-1], r2cv[i-1], msec[i-1], msecv[i-1] = pcr(X,y, pc=i)
```

```python
'''ç»˜åˆ¶ç”¨äºŽæ ¡å‡†å’Œäº¤å‰éªŒè¯ï¼ˆCVï¼‰çš„MSEé˜µåˆ—ã€‚
è¿½æ±‚çš„æ˜¯RMSE-CVçš„æœ€å°å€¼ï¼Œå³å‡æ–¹æ ¹è¯¯å·®ï¼ˆå³MSEçš„å¹³æ–¹æ ¹ï¼‰'''
with plt.style.context(('ggplot')):
    plt.plot(pc, np.sqrt(msecv[:]), 'r', label = "RMSECV")
    plt.plot(pc, np.sqrt(msec[:]), 'b', label = "RMSEC")
    plt.xlabel("# Principal Components")
    plt.ylabel("RMSE")
    plt.title("Conventional PCR")
    plt.legend()
    plt.show()
```

```python
predicted, r2r, r2cv, mser, mscv = pcr(X,y, pc=8)
# Regression plot
z = np.polyfit(y, predicted, 1)
with plt.style.context(('ggplot')):
    fig, ax = plt.subplots(figsize=(9, 5))
    ax.scatter(y, predicted, c='red', edgecolors='k')
    ax.plot(y, z[1]+z[0]*y, c='blue', linewidth=1)
    ax.plot(y, y, color='green', linewidth=1)
    plt.title('Conventional PCR -- $R^2$ (CV): {0:.2f}'.format(r2cv))
    plt.xlabel('Measured $^{\circ}$Brix')
    plt.ylabel('Predicted $^{\circ}$Brix')
    plt.show()
```
